# Example 1: Montecarlo Method

```python
from pyspark import SparkContext, SparkConf
import random
```

```python
# conf = SparkConf().setMaster("spark://spark-master:7075").setAppName("CircularEntropy")
conf = SparkConf().setMaster("local[6]").setAppName("CircularEntropy")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
 random.random()
```

```python
num_samples = 10000000
def inside(p):
    x, y = random.random(), random.random()
    return x*x + y*y < 1

samples = sc.parallelize(range(0, num_samples), 4)

# samples.collect()
```

```python
%%time 
count = samples.filter(inside).count()

pi = 4 * count / num_samples

pi
# samples.glom().collect()
```

# Example 2: WordCount

```python
LOCAL_DATA_PATH = '../../datasets/raw'
CLIENT_DATA_PATH = '../datasets'
tf = sc.textFile(f'{CLIENT_DATA_PATH}/words.txt',3)
```

```python
tf.take(1)
```

```python
tf.glom().collect()[0]
```

```python
def wordPairs(line):
    return line.split(' ')

tf2 = tf.flatMap(wordPairs)
tf2.collect()
tf3 = tf2.map(lambda word: (word, 1))
tf3.collect()

#def wordPairsReducer(e1, e2):
#    if e1[0] == e2[0]:
#        return (e1[0], e1[1]+e2[1])

# wordList = tf2.reduce(wordPairsReducer)
# wordList

counts = tf3.reduceByKey(lambda x,y: x+y)
```

```python
counts.collect()
```

# Store levels 
- Memory only: Almacenar RDD como objetos en memoria, si el RDD no cabe en la memoria lo restante no se almacena en disco. 
- Memory and disk: Almacena los RDDS como objetos de java no serializados en una JVM. Si el RDD no cabe en memoria, lo restante los va a alamacenar en disco pero va poder interactuar mas adelante cuando se necesite
